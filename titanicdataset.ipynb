**Uygulama 1:**
Cinsiyeti erkek, sınıfı ikinci sınıf ve mürettebat , yaşı 25’ten büyük  ve hayatta olan yolcuları  getiren sorguyu yazınız.

```
titanic[
    (titanic.sex == 'male')
    & (titanic['class'].isin(['Second', 'Crew']))
    & (titanic.age > 25)
    & (titanic.survived == 1)
]
```
